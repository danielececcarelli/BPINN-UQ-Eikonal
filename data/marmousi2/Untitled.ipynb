```python
import numpy as np
import matplotlib.pylab as plt

x = np.load("x.npy")
y = np.load("y.npy")
v = np.load("v.npy")
at = np.load("at.npy")
```

```python
np.max(x)
```

```python
plt.figure(figsize=(20,10))
plt.subplot()
plt.subplot(2,1,1)
plt.scatter(x, y, c=v, label = 'v', cmap = 'coolwarm', vmin = min(v), vmax = max(v))
plt.colorbar()
plt.xlim([0,np.max(x)])
plt.ylim([0,np.max(y)])


plt.subplot(2,1,2)
plt.scatter(x, y, c=at, label = 'at', cmap = 'coolwarm', vmin = min(at), vmax = max(at))
plt.colorbar()
plt.xlim([0,np.max(x)])
plt.ylim([0,np.max(y)])
```
